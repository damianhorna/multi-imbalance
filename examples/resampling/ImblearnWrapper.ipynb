Unzip datasets and prepare data:

```python
from collections import Counter

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA

from multi_imbalance.datasets import load_datasets
from multi_imbalance.utils.data import construct_flat_2pc_df

%matplotlib inline
sns.set_style("darkgrid")

df = load_datasets()["new_ecoli"]
X, y = df["data"], df["target"]
print(X[:5])
print(y[:5])
```

Oversample data using imblearn SMOTE for binary problems by using ImblearnWrapper

```python
from multi_imbalance.resampling.imblearn_wrapper import ImblearnWrapper
from imblearn.over_sampling import SMOTE

clf = ImblearnWrapper(SMOTE())
resampled_X, resampled_y = clf.fit_resample(X, y)
```

Compare results by plotting data in 2 dimensions

```python
n = len(Counter(y).keys())
p = sns.color_palette("husl", n)

pca = PCA(n_components=2)
pca.fit(X)

fig, axs = plt.subplots(ncols=2, nrows=2)
fig.set_size_inches(16, 10)
axs = axs.flatten()

axs[1].set_title("Base")
sns.countplot(x=y, ax=axs[0], palette=p)
X = pca.transform(X)
df = construct_flat_2pc_df(X, y)
sns.scatterplot(
    x="x1",
    y="x2",
    hue="y",
    style="y",
    data=df,
    alpha=0.7,
    ax=axs[1],
    legend="full",
    palette=p,
)


axs[3].set_title("Wrapped SMOTE")
sns.countplot(x=resampled_y, ax=axs[2], palette=p)
resampled_X = pca.transform(resampled_X)
df = construct_flat_2pc_df(resampled_X, resampled_y)
sns.scatterplot(
    x="x1",
    y="x2",
    hue="y",
    style="y",
    data=df,
    alpha=0.7,
    ax=axs[3],
    legend="full",
    palette=p,
)
```

Undersample data using imblearn ClusterCentroids for binary problems by using ImblearnWrapper

```python
from imblearn.under_sampling import ClusterCentroids

clf = ImblearnWrapper(ClusterCentroids())
resampled_X, resampled_y = clf.fit_resample(X, y)
```

Compare results by plotting data in 2 dimensions

```python
n = len(Counter(y).keys())
p = sns.color_palette("husl", n)

pca = PCA(n_components=2)
pca.fit(X)

fig, axs = plt.subplots(ncols=2, nrows=2)
fig.set_size_inches(16, 10)
axs = axs.flatten()

axs[1].set_title("Base")
sns.countplot(x=y, ax=axs[0], palette=p)
X = pca.transform(X)
df = construct_flat_2pc_df(X, y)
sns.scatterplot(
    x="x1",
    y="x2",
    hue="y",
    style="y",
    data=df,
    alpha=0.7,
    ax=axs[1],
    legend="full",
    palette=p,
)


axs[3].set_title("Wrapped SMOTE")
sns.countplot(x=resampled_y, ax=axs[2], palette=p)
resampled_X = pca.transform(resampled_X)
df = construct_flat_2pc_df(resampled_X, resampled_y)
sns.scatterplot(
    x="x1",
    y="x2",
    hue="y",
    style="y",
    data=df,
    alpha=0.7,
    ax=axs[3],
    legend="full",
    palette=p,
)
```

Oversample data using imblearn SMOTE for binary problems by using ImblearnWrapper with "static" mode, which works the same way as StaticSMOTE

```python
from multi_imbalance.resampling.static_smote import StaticSMOTE

clf = ImblearnWrapper(SMOTE(), mode="static")
resampled_X, resampled_y = clf.fit_resample(X, y)

clf = StaticSMOTE()
static_X, static_y = clf.fit_resample(X, y)
```

Compare results by plotting data in 2 dimensions

```python
n = len(Counter(y).keys())
p = sns.color_palette("husl", n)

pca = PCA(n_components=2)
pca.fit(X)

fig, axs = plt.subplots(ncols=2, nrows=3)
fig.set_size_inches(16, 15)
axs = axs.flatten()

axs[1].set_title("Base")
sns.countplot(x=y, ax=axs[0], palette=p)
X = pca.transform(X)
df = construct_flat_2pc_df(X, y)
sns.scatterplot(
    x="x1",
    y="x2",
    hue="y",
    style="y",
    data=df,
    alpha=0.7,
    ax=axs[1],
    legend="full",
    palette=p,
)


axs[3].set_title("Wrapped SMOTE in static mode")
sns.countplot(x=resampled_y, ax=axs[2], palette=p)
resampled_X = pca.transform(resampled_X)
df = construct_flat_2pc_df(resampled_X, resampled_y)
sns.scatterplot(
    x="x1",
    y="x2",
    hue="y",
    style="y",
    data=df,
    alpha=0.7,
    ax=axs[3],
    legend="full",
    palette=p,
)

axs[5].set_title("StaticSMOTE")
sns.countplot(x=static_y, ax=axs[4], palette=p)
resampled_X = pca.transform(static_X)
df = construct_flat_2pc_df(resampled_X, static_y)
sns.scatterplot(
    x="x1",
    y="x2",
    hue="y",
    style="y",
    data=df,
    alpha=0.7,
    ax=axs[5],
    legend="full",
    palette=p,
)
```
